```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('CO2 Emissions_Canada.csv')
data.head()
```

```python
data['Make'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
data['Make']=lb.fit_transform(data['Make'])
```

```python
data['Model']=lb.fit_transform(data['Model'])
```

```python
data['Vehicle Class']=lb.fit_transform(data['Vehicle Class'])
```

```python
data['Transmission']=lb.fit_transform(data['Transmission'])
```

```python
data['Fuel Type']=lb.fit_transform(data['Fuel Type'])
```

```python
data.info()
```

```python
data.isnull().sum()/len(data)*100
```

```python
from sklearn.model_selection import train_test_split
```

```python
TargetVariable=['CO2 Emissions(g/km)']
Predictors=['Make','Model','Vehicle Class','Engine Size(L)','Cylinders','Transmission','Fuel Type','Fuel Consumption City (L/100 km)',
            'Fuel Consumption Hwy (L/100 km)','Fuel Consumption Comb (L/100 km)','Fuel Consumption Comb (mpg)']
```

```python
X=data[Predictors].values
y=data[TargetVariable].values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
PredictorScaler=StandardScaler()
TargetVarScaler=StandardScaler()
```

```python
PredictorScalerFit=PredictorScaler.fit(X)
TargetVarScalerFit=TargetVarScaler.fit(y)
```

```python
X=PredictorScalerFit.transform(X)
y=TargetVarScalerFit.transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units=5, input_dim=11, kernel_initializer='normal', activation='relu'))
model.add(Dense(units=5, kernel_initializer='normal', activation='tanh'))
model.add(Dense(1, kernel_initializer='normal'))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train, y_train ,batch_size = 20, epochs = 50, verbose=1)
```

```python
Predictions=model.predict(X_test)
```

```python
Predictions=TargetVarScalerFit.inverse_transform(Predictions)
```

```python
y_test_orig=TargetVarScalerFit.inverse_transform(y_test)
```

```python
Test_Data=PredictorScalerFit.inverse_transform(X_test)
```

```python
TestingData=pd.DataFrame(data=Test_Data, columns=Predictors)
TestingData['CO2 Emissions(g/km)']=y_test_orig
TestingData['PredictedValue']=Predictions
TestingData.head()
```

```python
APE=100*(abs(TestingData['CO2 Emissions(g/km)']-TestingData['PredictedValue'])/TestingData['CO2 Emissions(g/km)'])
TestingData['APE']=APE

print('The Accuracy of ANN model is:', 100-np.mean(APE))
TestingData.head()
```
